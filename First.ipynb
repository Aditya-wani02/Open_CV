# Import Dependencies

```python
import os
import cv2
```

# Read in image

```python
img_path = os.path.join('spider.png')
```

```python
img_path
```

```python
img = cv2.imread(img_path)
```

```python
height , width , channels = img.shape
```

```python
height
```

# Render Image

```python
from matplotlib import pyplot as plt
```

```python
plt.imshow(img)
# it is showing diffrent color
```

```python
recolor =cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
plt.imshow(recolor)
```

```python
gray =cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
```

# Write out the file

```python
cv2.imwrite('grayimage.jpg' , gray)
```
